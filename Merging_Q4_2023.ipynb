## This notebook contains code to merge data scraped from the Turo website at two different time periods, preparing it for analysis.

---

```python
# Import relevant packages
import pandas as pd
```

---

### Import clean data and oct2023 data with jan2024 data

```python
# Read in both excel files
oct2023 = pd.read_excel(r"F:\Data Scraping\Turo\Nashville\Octorber\Clean Data\clean_data.xlsx")
jan2024 = pd.read_excel(r"F:\Data Scraping\Turo\Nashville\Jan 2024\clean data\clean data.xlsx")

# Merge files using an outer join
q4_data = pd.merge(oct2023, jan2024, on = 'id', how = 'outer')

# View the data
q4_data.head()
```

---

### Create several new columns
* `trips_Q4_2023` - Number of trips taken in the fourth quarter of 2023
* `price_increase_Q4_2023` - The price increase of each vehicle in the fourth quarter of 2023
* `rating_increase_Q4_2023` - The rating increase of each vehicle in the fourth quarter of 2023

```python
# create a new column for number of trips taken in Q4
q4_data['trips_Q4_2023'] = q4_data['trips_jan2024'] - q4_data['trips_oct2023']

# create a new column for the price increase in Q4
q4_data['price_increase_Q4_2023'] = q4_data['average_price_jan2024'] - q4_data['average_price_oct2023']

# create a new column for the rating increase in Q4
q4_data['rating_increase_Q4_2023'] = q4_data['average_rating_jan2024'] - q4_data['average_rating_oct2023']

# view the data
q4_data.head()
```

---

### Export to excel

```python
q4_data.to_excel(r"D:\Data Scraping\Turo\Nashville\clean data\q4_2023_data.xlsx", index = False)
```
